### Checking the best suitable model for prediction.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25)
```

```python

from sklearn import metrics
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
```

```python
def predict_me(my_model):
    model = my_model()
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    print('The accuracy of the model is : ', metrics.r2_score(y_test, prediction))
    print('mean absolute error is : ', metrics.mean_absolute_error(y_test, prediction))
  
# tell why you are taking r2 score mean absolute error(at every prediction we have error of appx 1800 /-)
```

```python
predict_me(LinearRegression )
```

```python
predict_me(KNeighborsRegressor)
```

```python
predict_me(DecisionTreeRegressor)
```

```python
#we can see that as efficiency increases mean absolute error decreases.
```

```python
predict_me(RandomForestRegressor)
```

```python
predict_me(Lasso)
```

```python
predict_me(Ridge)
```

```python
predict_me( ElasticNet)
```

### Optimizing our model 

```python
# We will get best parameters and hence will get the best score for the respective model.
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.ensemble import RandomForestRegressor 
rs=RandomForestRegressor()
```

```python
n_est = [int(x) for x in np.linspace(start=100,stop=1000,num=7)]
max_dep = [int(x) for x in np.linspace(start=1,stop=60,num=4)]
```

```python
optimizer = {
'n_estimators' : n_est, 'max_depth' : max_dep,
 'max_features' : ['auto', 'sqrt'],
'min_samples_split' : [5, 10, 15, 100]
}
```

```python
opti_model = RandomizedSearchCV(estimator = rs, param_distributions = optimizer, cv=3, verbose=2, n_jobs=1)
```

```python
opti_model.fit(X_train, y_train)
```

```python
opti_model.best_params_
```

```python
predins = opti_model.predict(X_test)
```

```python
print(metrics.r2_score(y_test, predins))
#The efficiency has increased by hypertuning our models.
```
